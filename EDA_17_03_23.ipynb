## EDA

# UNI VARIATE ANALYSIS

### Hist plot

```python
sns.histplot(df.children , color= 'black')
plt.show()
```

### BOXPLOT

```python
sns.boxplot('age', data=df)
plt.show()
```

# MULTI VARIATE ANALYSIS

## COUNT PLOT

```python
plt.figure(figsize=(40,9))
sns.countplot(x='age' ,data=df)
plt.show()
```

```python
plt.figure(figsize=(4,2))
sns.countplot(x= 'region',data= df)
plt.show()
```

## Bar Plot

```python
plt.figure(figsize=(5,3))
sns.barplot(x='region', y='charges', hue='sex' , data=df, )
plt.show()
```

## Line Plot

```python
plt.figure(figsize= (5,3))
sns.lineplot(x= 'region', y= 'age', data=df ,hue='sex')
plt.show()
```

```python
sns.lineplot('age','bmi', data = df)
plt.show()
```

### Factor plolt

```python
sns.set_style('ticks')
sns.factorplot('region','age', hue='sex', data=df)
plt.show()
```

### PIE Chart

```python
re_se =df.groupby(['region'])['charges'].sum()
re_se
```

```python
sns.set_style('ticks')
re_se.plot.pie()
plt.legend( loc ='lower left')
plt.show()
```

## PAIR PLOT

```python
sns.pairplot(data= df, hue= 'sex')
plt.show()
```

```python
heat_map= df.corr()
sns.heatmap(heat_map)
plt.show()
```

### Violin plot

```python
sns.set_style('dark')
sns.violinplot('sex','age', hue ='region', data= df)
plt.show()
```
